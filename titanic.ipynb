```python
# erkaklar soni
male_count = df[df.Sex=='male'].shape[0]
# ayollar soni
female_count = df[df.Sex=='female'].shape[0]
```

```python
# eng: Survived male count
# uz: tirik qolgan erkaklar soni

male_survived = df[(df.Sex=='male') & (df.Survived==1)].shape[0]
```

```python
# eng: Survived female count
# uz: tirik qolgan ayollar soni

female_survived = df[(df.Sex=='female') & (df.Survived==1)].shape[0]
```

```python
# eng: male survived rate
# uz: erkaklarning omon qolish darajasi

male_survived_rate = (male_survived / male_count) * 100
male_survived_rate
```

```python
# eng: female survived rate
# uz: ayollarning omon qolish darajasi

female_survived_rate = (female_survived / female_count) * 100
female_survived_rate
```

```python
# erkaklarning omon qolish darajasini ayollarning omon qolish darajasi bilan ulush hisobi
plt.pie([male_survived_rate, female_survived_rate], labels=['Male survived rate', 'Female survived rate'])
```

```python
plt.pie
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Male', 'Female'
sizes = [male_count, female_count]
explode = (0.1, 0.1)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%.2f%%',
        shadow=True, startangle=30)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
